````python
from sklearn.model_selection import train_test_split
````

````python
df.columns #
````

````python
X = df[['Price (Rs.)', 'Discount (%)', 'Final_Price(Rs.)', 'Payment_Method']] #Dataframe de atributos básicos;
y = df['Category'] #Dataframe de atributos alvos;
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) #definindo a proporção de 30% para o conjunto de teste.
````

````python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
````

````python
[X_train, X_test, y_train, y_test] = train_test_split(X, y, test_size=0.2)
X_train
````

#> Treinando o modelo com os dados de treinamento

````python
# Importando bibliotecas necessarias
from sklearn.linear_model import LogisticRegression # Modelo exemplo
from sklearn.preprocessing import LabelEncoder, OneHotEncoder # Importando OneHotEncoder
from sklearn.compose import ColumnTransformer # Importando ColumnTransformer
from sklearn.metrics import r2_score # Importando r2_score

# Criando uma instancia de um modelo
model = LogisticRegression()
````

````python
# Criando um LabelEncoder
encoder = LabelEncoder()
````

````python
# Ajustando o codificador aos seus dados de treinamento e transfando-os
y_train = encoder.fit_transform(y_train)
y_test = encoder.fit_transform(y_test)
````

# Identificando a coluna categorica do DataFrame:

````python
categorical_cols = X_train.select_dtypes(include=['object']).columns.tolist()
````

# Criando um ColumnTransformer para aplicar OneHotEncoding a recursos categóricos:

````python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', X_train.select_dtypes(exclude=['object']).columns.tolist()),
        ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_cols)],
    remainder='drop'  # Drop other columns if needed
)
````

````python
# Ajustando e transformando os dados de treinamento
X_train_encoded = preprocessor.fit_transform(X_train)
X_test_encoded = preprocessor.fit_transform(X_test)

````

````python
# Ajustando o modelo
model.fit(X_train_encoded, y_train)
````

# Fazendo predições com os dados de teste:

````python
y_pred = model.predict(X_train_encoded)
y_pred
````

#> Calculando o R²

````python
# Importando a função necessaria de scikit-learn
from sklearn.metrics import r2_score
````

````python
# Prevendo com base nos dados de teste, não nos dados de treinamento
y_pred = model.predict(X_test_encoded)
````

````python
# Calculate R-squared
r2 = r2_score(y_test, y_pred)
r2
````
